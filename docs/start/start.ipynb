# QAOA for MaxCut

This notebook demonstrates how to use `qiskit.algorithms.minimum_eigensolvers.QAOA` to solve the MaxCut problem.

## Introduction
Given a graph $G=(V, E)$, the MaxCut problem seeks to partition the set of vertices $V$ into two subsets, $V_0$ and $V_1$, such that the number of edges between $V_0$ and $V_1$ is maximized. This is a classic NP-hard problem.

## MaxCut as a Quadratic Unconstrained Binary Optimization (QUBO) Problem
The MaxCut problem can be formulated as a QUBO problem. Let $x_i \in \{0, 1\}$ be a binary variable for each vertex $i \in V$, where $x_i = 0$ if vertex $i$ is in $V_0$ and $x_i = 1$ if vertex $i$ is in $V_1$. The objective function to maximize is the number of edges between $V_0$ and $V_1$. For an edge $(i, j) \in E$, if $x_i \neq x_j$, then the edge contributes to the cut. This can be expressed as $x_i(1-x_j) + x_j(1-x_i)$.

Alternatively, consider the value of an edge $(i,j)$ in the cut as $1/2(1 - z_i z_j)$ where $z_i \in \{-1, 1\}$. If $z_i = z_j$, the edge is not cut. If $z_i = -z_j$, the edge is cut. We want to maximize $\sum_{(i,j) \in E} 1/2(1 - z_i z_j)$.

By mapping $z_i \rightarrow 1 - 2x_i$, where $x_i \in \{0,1\}$, we can convert this to a binary formulation.

Maximizing the cut is equivalent to minimizing the negative of the cut objective function. The Hamiltonian for MaxCut can be written as:

$H = -\sum_{(i,j) \in E} \frac{1}{2} (1 - Z_i Z_j)$

```
import numpy as np
import networkx as nx
from qiskit_algorithms import QAOA
from qiskit_algorithms.optimizers import COBYLA
from qiskit.quantum_info import Pauli, SparsePauliOp
from qiskit.primitives.sampler import Sampler  # Modified import for Sampler
from qiskit_optimization.applications import Maxcut
from qiskit_optimization.converters import QuadraticProgramToQubo
```

## Graph Definition
Let's define a simple graph for the MaxCut problem.

```
# Generate a graph G with 4 nodes
n_nodes = 4
G = nx.Graph()
G.add_nodes_from(np.arange(0, n_nodes, 1))
# G.add_edges_from([(0, 1), (0, 2), (0, 3), (1, 2), (2, 3)])
G.add_edges_from([(0, 1), (1, 2), (2, 3), (3, 0), (0,2)]) # A cycle graph C4 with an extra edge

# Draw the graph
colors = ['r' for node in G.nodes()]
pos = nx.spring_layout(G)
nx.draw_networkx(G, node_color=colors, node_size=600, alpha=0.8, pos=pos)
print(f"Number of nodes: {n_nodes}")
print(f"Number of edges: {len(G.edges())}")
```

## Mapping MaxCut to an Ising Hamiltonian
The MaxCut problem can be mapped to an Ising Hamiltonian. The `Maxcut` class from `qiskit_optimization.applications` can help us with this.

```
# Create a Maxcut application instance
max_cut = Maxcut(G)

# To get the Ising Hamiltonian, we first need to convert to a QuadraticProgram
qp = max_cut.to_quadratic_program()
print("Quadratic Program:")
print(qp)

# Convert the QuadraticProgram to a QUBO
conv = QuadraticProgramToQubo()
qubo = conv.convert(qp)
print("\nQUBO:")
print(qubo)

# Convert QUBO to an Ising Hamiltonian
# The `to_ising()` method directly gives the operator and offset from the Maxcut application.
# The MaxCut application directly creates the Hamiltonian in the Ising format (Z operators).
qubit_op, offset = max_cut.to_ising()
print("\nIsing Hamiltonian (Qubit Operator):")
print(qubit_op)
print(f"Offset: {offset}")
```

## Running QAOA
Now we will use the `QAOA` algorithm to find the solution for MaxCut. We will use `COBYLA` as the optimizer and `Sampler` as the primitive.

```
# Set up the optimizer
optimizer = COBYLA()

# Set up the primitive
sampler = Sampler()

# Initialize QAOA
qaoa = QAOA(sampler=sampler, optimizer=optimizer, reps=1) # reps=1 for simplicity

# Run QAOA
result = qaoa.compute_minimum_eigenvalue(qubit_op)

print("QAOA Result:")
print(result)

# The result object contains the optimal parameters and the measurement outcomes.
# The eigenstate is given as a sparse pauli op, so we can convert to bitstring representation to see partitions
optimal_parameters = result.optimal_parameters
optimal_value = result.optimal_value
measurement_outcomes = result.eigenstate.binary_probabilities()

print(f"Optimal parameters: {optimal_parameters}")
print(f"Optimal value (eigenvalue): {optimal_value}")
print(f"Measurement outcomes: {measurement_outcomes}")
```

## Analyzing Results
Let's analyze the measurement outcomes to find the MaxCut partition.

```
# Get the most probable bit string
most_probable_bitstring = max(measurement_outcomes, key=measurement_outcomes.get)
print(f"Most probable bitstring: {most_probable_bitstring}")

# Convert the bitstring to a partition
partition = [int(bit) for bit in most_probable_bitstring]
print(f"Partition: {partition}")

# Calculate the cut value for this partition
cut_value = max_cut.get_quadratic_program().objective.evaluate(partition)
print(f"Cut value from partition: {-cut_value} (negated because we minimized -MaxCut)")
```

```
# Visualize the partition
colors = ['r' if partition[i] == 0 else 'b' for i in range(n_nodes)]
nx.draw_networkx(G, node_color=colors, node_size=600, alpha=0.8, pos=pos)

# Calculate the actual MaxCut value for verification
def calculate_maxcut_value(graph, partition_assignment):
    cut_count = 0
    for u, v in graph.edges():
        if partition_assignment[u] != partition_assignment[v]:
            cut_count += 1
    return cut_count

actual_max_cut = calculate_maxcut_value(G, partition)
print(f"Actual MaxCut value from graph calculation: {actual_max_cut}")
```
